```python
from faker import Faker
import pymysql
import random
import datetime

fake = Faker('en_IN')

def mobile_number():
    start = random.choice(['7','8','9'])
    remaining = ''.join(random.choices('0123456789',k=9))
    return start+remaining

def generate_random_datetime(start_year=2020, end_year=2023):

  year = random.randint(start_year, end_year)
  month = random.randint(1, 12)
  day = random.randint(1, 28)  # Adjust for leap years if needed

  hour = random.randint(0, 23)
  minute = random.randint(0, 59)
  second = random.randint(0, 59)

  random_datetime = datetime.datetime(year, month, day, hour, minute, second)
  return random_datetime.strftime('%Y-%m-%d %H:%M:%S')

try: 
    connection = pymysql.connect(
        host="localhost",
        user="root",
        password="mysql",
        database="zomata",
    )
    cursor = connection.cursor()
    print("Connection established successfully")

    insert_restaurant = """insert into restaurant(restaurant_id,name,cuisine_type,location,owner_name,average_delivery_time_min,contact_number,rating,total_orders,is_active)
    values(%s,%s,%s,%s,%s,%s,%s,%s,%s,%s)"""

    restaurant = []
    for i in range(50):
        restaurant.append((
            i+1,
            fake.company(),
            fake.word(ext_word_list=['Indian','Chinese','Italian','Mexican','Thai','Japanese']),
            fake.city(),
            fake.name(),
            random.randint(10,60),
            mobile_number(),
            round(random.uniform(1.0,5.0),1),
            random.randint(0,500),
            fake.boolean(chance_of_getting_true=80)
        ))

    cursor.executemany(insert_restaurant,restaurant)
    connection.commit()
    print("Restaurant table data populated successfully")

    insert_customer = """insert into customer(customer_id,name,email,phone,location,signup_date,is_premium,preferred_cuisine,total_orders,average_rating)
    values(%s,%s,%s,%s,%s,%s,%s,%s,%s,%s)"""

    customer = []
    for i in range(100):
        customer.append((
            i+1,
            fake.name(),
            fake.email(),
            mobile_number(),
            fake.city(),
            fake.date_this_decade(),
            fake.boolean(chance_of_getting_true=30),
            fake.word(ext_word_list=['Indian','Chinese','Italian','Mexican','Thai','Japanese']),
            random.randint(0,100),
            round(random.uniform(1.0,5.0),1)
        ))

    cursor.executemany(insert_customer,customer)
    connection.commit()
    print("Customer data populated successfully")

    insert_orders = """insert into orders(order_id, customer_id, restaurant_id, order_date, delivery_time, status, total_amount, payment_mode, discount_applied, feedback_rating)
    VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s)"""

    order = []
    for i in range(1000):
        order.append((
            i+1,
        	random.randint(1,100),
            random.randint(1,30),
            generate_random_datetime(),
            generate_random_datetime(),
            random.choice(['pending','delivered','cancelled','inprogress']),
            round(random.uniform(100.0,9999.0),2),
            random.choice(['card','cash','upi','netbanking']),
            round(random.uniform(20.0,200.0),2),
            round(random.uniform(1.0,5.0),1)
        ))

    cursor.executemany(insert_orders,order)
    connection.commit()
    print("Order data populated successfully")

    insert_delivery_person = """insert into delivery_person(delivery_person_id, name, contact_number, total_deliveries, average_rating, location)
    VALUES (%s, %s, %s, %s, %s, %s)"""

    delivery_person = []
    for i in range (30):
        delivery_person.append((
            i+1,
            fake.name(),
            mobile_number(),
            random.randint(0,100),
            round(random.uniform(1.0,5.0),1),
            fake.city()
        ))

    cursor.executemany(insert_delivery_person,delivery_person)
    connection.commit()
    print("Delivery person data updated successfully")

    insert_delivery = """insert into delivery(delivery_id, order_id, delivery_person_id, delivery_status, distance_in_km, delivery_time_min, estimated_time_min, delivery_fee, vehicle_type)
    VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s)"""

    delivery = []
    for i in range(1000):
        delivery.append((
            i+1,
            random.randint(1,1000),
            random.randint(1,30),
            random.choice(['pending','delivered','cancelled','inprogress']),
            random.randint(1,30),
            random.randint(15,60),
            random.randint(12,48),
            round(random.uniform(5.0,50.0),1),
            random.choice(['Bike','Car','Cycle'])
        ))

    cursor.executemany(insert_delivery,delivery)
    connection.commit()
    print("Delivery data updated successfully")

except pymysql.MySQLError as e:
    print(f"Error:{e}")
finally:
    if connection:
        cursor.close()
        connection.close()
        print("connection closed")
```
